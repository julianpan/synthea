```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
sns.set()
%matplotlib inline
```

```python
states = pd.read_csv('drive/MyDrive/2021 HHS Data Challenge/Data/states.csv', header=None, names=['name'])
states.head(3)
```

```python
run_name = 'sample_10k_diabetes_education'
```

```python
all_patients = None

for state in states.name.values:
  data_path = 'drive/MyDrive/2021 HHS Data Challenge/Data/'+run_name+'/'+state+'/csv/patients.csv.gz'
  print(data_path)
  state_df = pd.read_csv(data_path)
  if all_patients is None:
    all_patients = state_df
  else:
    all_patients = all_patients.append(state_df)
```

```python
now = pd.Timestamp('now')
all_patients['AGE']=(now - pd.to_datetime(all_patients.BIRTHDATE)).astype('<m8[Y]') 
all_patients['ALIVE']=all_patients.DEATHDATE.isna()
```

```python
all_condition_df = None

for state in states.name.values:
  data_path = 'drive/MyDrive/2021 HHS Data Challenge/Data/'+run_name+'/'+state+'/csv/conditions.csv.gz'
  print(data_path)
  condition_df = pd.read_csv(data_path)
  if all_condition_df is None:
    all_condition_df = condition_df
  else:
    all_condition_df = all_condition_df.append(condition_df)
```

```python
all_observations_df = None

for state in states.name.values:
  data_path = 'drive/MyDrive/2021 HHS Data Challenge/Data/'+run_name+'/'+state+'/csv/observations.csv.gz'
  print(data_path)
  observation_df = pd.read_csv(data_path)
  if all_observations_df is None:
    all_observations_df = observation_df
  else:
    all_observations_df = all_observations_df.append(observation_df)
```

```python
prediabetes_start = all_condition_df[all_condition_df.DESCRIPTION == 'Prediabetes'][['PATIENT', 'START']].rename(columns={'PATIENT': 'Id', 'START': 'Prediabetes_START'}).reset_index(drop=True)
diabetes_start = all_condition_df[all_condition_df.DESCRIPTION == 'Diabetes'][['PATIENT', 'START']].rename(columns={'PATIENT': 'Id', 'START': 'Diabetes_START'}).reset_index(drop=True)
obesity_start = all_condition_df[all_condition_df.DESCRIPTION.str.contains('Body mass index 30+')][['PATIENT', 'START']].rename(columns={'PATIENT': 'Id', 'START': 'Obesity_START'}).reset_index(drop=True)
severely_obesity_start = all_condition_df[all_condition_df.DESCRIPTION.str.contains('Body mass index 40+')][['PATIENT', 'START']].rename(columns={'PATIENT': 'Id', 'START': 'Severely_Obesity_START'}).reset_index(drop=True)
a1c_max = all_observations_df[all_observations_df.CODE=='4548-4'].groupby('PATIENT')[['VALUE']].max().reset_index().rename(columns={'PATIENT': 'Id', 'VALUE': 'A1c_max'})
all_patients['Prediabetes'] = all_patients.Id.isin(prediabetes_start['Id'])
all_patients['Diabetes'] = all_patients.Id.isin(diabetes_start['Id'])
all_patients['Obesity'] = all_patients.Id.isin(obesity_start['Id'])
all_patients['Severely_Obesity'] = all_patients.Id.isin(severely_obesity_start['Id'])
merged_df = all_patients.merge(prediabetes_start, on='Id', how='left').merge(diabetes_start, on='Id', how='left').merge(obesity_start, on='Id', how='left').merge(a1c_max, on='Id', how='left')

```
